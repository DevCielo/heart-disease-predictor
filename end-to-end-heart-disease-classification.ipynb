## Modelling

```python
df.head()
```

```python
# Split data into x and y
X = df.drop("target", axis=1)
y = df["target"]

np.random.seed(42)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
X_train
```

```python
models = {"Logistic Regression": LogisticRegression(),
          "KNN": KNeighborsClassifier(),
          "RandomForest": RandomForestClassifier()}

def fit_and_score(models, X_train, X_test, y_train, y_test):
    np.random.seed(42)
    models_scores = {}
    for name, model in models.items():
        model.fit(X_train, y_train)
        models_scores[name] = model.score(X_test, y_test)
        
    return models_scores
```

```python
model_scores = fit_and_score(models=models, X_train=X_train, X_test=X_test, y_train=y_train, y_test=y_test)

model_scores
```

```python
model_compare = pd.DataFrame(model_scores, index=["accuracy"])
model_compare.T.plot.bar();
```

# Hyperparamater Tuning

```python
train_scores = []
test_scores = []

# Create a list of different numbers for n_neighbors
neighbors = range(1,21)

knn = KNeighborsClassifier();

for i in neighbors:
    knn.set_params(n_neighbors=i)
    
    knn.fit(X_train, y_train)
    
    train_scores.append(knn.score(X_train, y_train))
    
    test_scores.append(knn.score(X_test, y_test))

```

```python
train_scores[:10]
```

```python
test_scores[:10]
```

```python
plt.plot(neighbors, train_scores, label="Train score")
plt.plot(neighbors, test_scores, label="Test score")
plt.xticks(range(1,21,1))

plt.title("Train versus Test Score")
plt.ylabel("Score")
plt.xlabel("Neighbors")
plt.legend(["Train", "Test"]);

print(max(test_scores))
```

# Hyperparamater tuning with RandomizedSearchCV

```python
log_reg_grid = {"C": np.logspace(-4, 4, 20),
                "solver": ["liblinear"]}

rf_grid = {"n_estimators": np.arange(10, 1000, 50),
          "max_depth": [None, 3, 5, 10],
          "min_samples_split": np.arange(2, 20, 2),
          "min_samples_leaf": np.arange(1, 20, 2)}
```

```python
# Tune LogisticRegression
np.random.seed(42)

rs_log_reg = RandomizedSearchCV(LogisticRegression(), param_distributions=log_reg_grid, cv=5, n_iter=20, verbose=True)

rs_log_reg.fit(X_train, y_train)
```

```python
rs_log_reg.best_params_
```

```python
rs_log_reg.score(X_test, y_test)
```

```python
np.random.seed(42)

rs_rf = RandomizedSearchCV(RandomForestClassifier(), param_distributions=rf_grid, cv=5, n_iter=20, verbose=True)

rs_rf.fit(X_train, y_train)
```

```python
rs_rf.best_params_
```

```python
rs_rf.score(X_test, y_test)
```

### Hyperparamater tuning with GridSearchCV

```python
np.random.seed(42)
log_reg_grid = {"C": np.logspace(-4, 4, 30),
                "solver": ["liblinear"]}

gs_log_reg = GridSearchCV(LogisticRegression(), param_grid=log_reg_grid, cv=5, verbose=True)

gs_log_reg.fit(X_train, y_train)
```

```python
gs_log_reg.score(X_train, y_train)
```

```python
gs_log_reg.best_params_
```

## Evaluating our tuned machine learning model

```python
y_preds = gs_log_reg.predict(X_test)
```

```python
RocCurveDisplay.from_estimator(gs_log_reg, X_test, y_test)
plt.plot([0,1],[0,1],"k--");
```

```python
print(confusion_matrix(y_test, y_preds))
```

```python
sns.set(font_scale=1.5)

def plot_conf_mat(y_test, y_preds):
    fig, ax = plt.subplots(figsize=(3,3))
    ax = sns.heatmap(confusion_matrix(y_test, y_preds),
                     annot=True, #the numbers in the boxes
                     cbar=False) #Removes the colour bar
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    
plot_conf_mat(y_test, y_preds)
```

```python
print(classification_report(y_test, y_preds))
```
